```python
import pandas as pd
```

```python
hack = pd.read_csv("./data/housing_hack.csv")
test = pd.read_csv("./data/houseprice_test.csv")
```

```python
hack.head()
hack["Population"] = [float(i) for i in hack.Population]
hack["HouseAge"] = [float(i) for i in hack.HouseAge]
```

```python
row = test.iloc[10,:]
for i in row:
    print(i)
```

```python
def match(test_row,df,cols):
    print(test_row)
    for i in range(len(test_row)):
        df = df[abs(df[cols[i]]-float(test_row[i]))<=0.0001]
        if len(df)==1:
            break
    if len(df)>1:
        print("there may be not a correct answer!!!!")
    print(df["target"])
    return df["target"]
```

```python
hack_answer = []
for i in range(len(test)):
    hack_answer.append(match(test.iloc[i,:],df=hack,cols =list(hack.columns)).values[0])
```

```python
hack_answer
```

```python
hack_ = pd.DataFrame(hack_answer)
```
